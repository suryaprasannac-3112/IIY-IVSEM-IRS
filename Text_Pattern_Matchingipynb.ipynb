```python
def locate(main_string, sub_string):
  size_main = len(main_string)
  size_sub = len(sub_string)
  print("given string:", main_string)
  print("sub string:", sub_string)
  print("-----------------------------------------------")
  found_any = 0
  index = 0
  while index <= size_main - size_sub:
    k = 0
    while k < size_sub:
      if main_string[index + k] != sub_string[k]:
        break
      k = k + 1
    if k == size_sub:
      print("substring exist at postion", index)
      found_any = 1
    index = index + 1
  if found_any == 0:
    print("substring not found in given string")

full_text = "AABAAACADAABAABA"
search_text = "AAABA"
locate(full_text, search_text)
```
